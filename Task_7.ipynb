1.Perform Bivariate analysis using the pandas DataFrame that contains information about two variables: (1) Hours spent studying and (2) Exam score received by 20 different students:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data
data = {'Hours_Studied': [5, 8, 2, 10, 7, 6, 3, 4, 9, 1, 8, 5, 7, 6, 2, 10, 3, 9, 4, 1],
        'Exam_Score': [60, 85, 30, 95, 75, 70, 40, 50, 90, 20, 85, 65, 80, 78, 35, 92, 42, 88, 55, 25]}

df = pd.DataFrame(data)

# Scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Hours_Studied', y='Exam_Score', data=df)
plt.title('Scatter Plot of Hours Studied vs Exam Score')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.show()

# Correlation coefficient
correlation = df['Hours_Studied'].corr(df['Exam_Score'])
print(f'Correlation coefficient: {correlation}')

# Regression plot
plt.figure(figsize=(8, 6))
sns.regplot(x='Hours_Studied', y='Exam_Score', data=df)
plt.title('Regression Plot of Hours Studied vs Exam Score')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.show()
```

 2. i. Using Pandas, Create a DataFrame with a list of dictionaries, row indices, and column
indices.
ii. Use index label to delete or drop rows from a Pandas DataFrame.

```python
import pandas as pd

# Sample data
data = [
    {'Name': 'John', 'Age': 25, 'City': 'New York'},
    {'Name': 'Alice', 'Age': 30, 'City': 'Los Angeles'},
    {'Name': 'Bob', 'Age': 22, 'City': 'Chicago'},
    {'Name': 'Eva', 'Age': 28, 'City': 'San Francisco'},
]

# Creating a DataFrame
df = pd.DataFrame(data, index=['A', 'B', 'C', 'D'])

# Displaying the DataFrame
print("DataFrame:")
print(df)
```

```python
# Deleting a row by index label
index_to_delete = 'C'
df = df.drop(index_to_delete)

# Displaying the DataFrame after deletion
print(f"\nDataFrame after deleting row with index '{index_to_delete}':")
print(df)
```

4.i. Write a NumPy program to convert a list and tuple into arrays
ii.Write a NumPy program to convert the values of Centigrade degrees into Fahrenheit
degrees and vice versa. Values have to be stored into a NumPy array.

```python
import numpy as np

# Convert a list to a NumPy array
list_data = [1, 2, 3, 4, 5]
array_from_list = np.array(list_data)

# Convert a tuple to a NumPy array
tuple_data = (6, 7, 8, 9, 10)
array_from_tuple = np.array(tuple_data)

# Display the arrays
print("Array from List:", array_from_list)
print("Array from Tuple:", array_from_tuple)
```

```python
import numpy as np

def celsius_to_fahrenheit(celsius_values):
    return np.array(celsius_values) * 9/5 + 32

def fahrenheit_to_celsius(fahrenheit_values):
    return (np.array(fahrenheit_values) - 32) * 5/9

# Example values in Celsius
celsius_temperatures = [0, 10, 20, 30, 40]

# Convert Celsius to Fahrenheit
fahrenheit_temperatures = celsius_to_fahrenheit(celsius_temperatures)

# Display the conversion
print("Celsius Temperatures:", celsius_temperatures)
print("Fahrenheit Temperatures:", fahrenheit_temperatures)

# Example values in Fahrenheit
fahrenheit_temperatures_2 = [32, 50, 68, 86, 104]

# Convert Fahrenheit to Celsius
celsius_temperatures_2 = fahrenheit_to_celsius(fahrenheit_temperatures_2)

# Display the conversion
print("\nFahrenheit Temperatures 2:", fahrenheit_temperatures_2)
print("Celsius Temperatures 2:", celsius_temperatures_2)
```
