```python
# guess prints what
x = 88
def f():
    x = 21
    return x
print(x)      # x = 88 global scope
print(f())    # x = 21 local scope
```

```python
# What if there is no local scope
x = 88
def f():
    y = 21*x        # there is no local scope x
    return y
print(f())         # it uses global scope x
# First local scope searched, then global scope searched, if two of them cannot be found lastly built in scope searched.
```

```python
#nested function
def square():
    """ return square of value """
    def add():
        """ add two local variable """
        x = 22
        y =88
        z = x + y
        return z
    return add()**2
print(square())
```

```python
# default arguments
def f(a, b = 1, c = 2):
    y = a + b + c
    return y
print(f(5))
# what if we want to change default arguments
print(f(5,4,3))
```

```python
# flexible arguments *args
def f(*args):
    for i in args:
        print(i)
f(1)
print("")
f(1,2,3,4)
# flexible arguments **kwargs that is dictionary
def f(**kwargs):
    """ print key and value of dictionary"""
    for key, value in kwargs.items():# If you do not understand this part turn for loop part and look at dictionary in for loop
        print(key, " ", value)
f(players = 'Virat kohli', runs = '88', strike_rate ="93.61" )
```

```python
# zip example
list1 = [1,2,3,4]
list2 = [5,6,7,8]
z = zip(list1,list2)
print(z)
z_list = list(z)
print(z_list)
```

```python
# Example of list comprehension
num1 = [1,2,3]
num2 = [i + 1 for i in num1 ]
print(num2)
```

```python
# Conditionals on iterable
num1 = [5,10,15]
num2 = [i**2 if i == 10 else i-5 if i < 7 else i+5 for i in num1]
print(num2)
```
